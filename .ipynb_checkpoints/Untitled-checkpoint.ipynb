```python
#Keep Relevant Functions Here

def wholeDatabaseToPandas(database = "cv4DatabaseSarMdac.db", table = "cv4"):
    conn = sqlite3.connect(database)
    c = conn.cursor()
    df = DataFrame(c.execute(f"SELECT * FROM {table}").fetchall())

    column_names = []
    for column in c.execute(f"pragma table_info({table});").fetchall():
        column_names.append(column[1])
    df.columns = column_names

    conn.commit()
    conn.close()

    return df

def getSpecificChannelDF(chNum, database = "cv4DatabaseSarMdac.db", table = "cv4"):
    constantsDF = wholeDatabaseToPandas(database, table)
    channelDF = constantsDF.loc[constantsDF['channel'] == chNum]
    return channelDF
```
